    # Standardizing the features
    scaler = StandardScaler()
    X = scaler.fit_transform(iris_df.drop('species', axis=1))  # Ensure the column name is correct

    # Model Training
    kmeans = KMeans(n_clusters=3, random_state=42)
    kmeans_labels = kmeans.fit_predict(X)
    iris_df['KMeans_Cluster'] = kmeans_labels

    hierarchical = AgglomerativeClustering(n_clusters=3)
    hierarchical_labels = hierarchical.fit_predict(X)
    iris_df['Hierarchical_Cluster'] = hierarchical_labels

    dbscan = DBSCAN(eps=0.5, min_samples=5)
    dbscan_labels = dbscan.fit_predict(X)
    iris_df['DBSCAN_Cluster'] = dbscan_labels
